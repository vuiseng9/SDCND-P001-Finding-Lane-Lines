```python
import cv2
print(cv2.__version__)
vidcap = cv2.VideoCapture('./test_videos/challenge.mp4')
success,image = vidcap.read()
count = 0
success = True
while success:
    success,image = vidcap.read()
    print(str(count), 'Read a new frame: ', success)
    cv2.imwrite("./challenge_images/challenge_img_%d.jpg" % count, image)     # save frame as JPEG file
    count += 1
```
